# On Reproducibility of LLM-generated Respondents (2): Varied Random Seeds

In this notebook, we want to verify whether the LLM-generated respondents are reproducible across different random seeds.

## Experiment Settings

The experimental settings are the same as the previous one:

- LLM: `gemma4-e4b`
- Questionnaire: `bigfive50.csv`
- Prompt: `persona + basic_instruction + item + format_instruction`.
    - Basic instruction:```Indicate for the statement whether it is 1. Very Inaccurate, 2. Moderately Inaccurate, 3. Neither Accurate Nor Inaccurate, 4. Moderately Accurate, or 5. Very Accurate as a description of you.```
    - Format instruction:```Output the results as a json object: {'rating':'1-5' ,'reasoning':'a simple reasoning of the rating in one sentence'}```
    - Persona:```You are a 35-year0old male. You are a senior software engineer at Google. You are married and you have a 2-year-old girl child.```
- Random Seed: 100 numbers from `10101` to `10200`
- `temperature` is set to `0.0`

## Load Experimental Data

````python
## Load CSV data
import os
import pandas as pd

DATAPATH = '../output/rseed/'
files = os.listdir(DATAPATH)
print('Totally '+str(len(files))+' files.')
print('Read the first file: '+DATAPATH+files[0])
df = pd.read_csv(DATAPATH+files[0], index_col=0)
print(df.shape)
print(df.head())

## Separating rating and reasoning
ratings = []
reasonings = []
#
for f in files:
    tmp = pd.read_csv(DATAPATH+f, index_col=0)
    ratings.append(tmp['rating'])
    reasonings.append(tmp['reasoning'])

#
df_ratings= pd.concat(ratings, axis=1).transpose()
df_reasoning = pd.concat(reasonings, axis=1).transpose()
````

## Variability of Responses

````python
df_ratings.std()
````

As shown above, there is no variability across the 100 tests; hence, the LLM-generated responses are reproducible with a fixed prompt and varied random seeds. Note that we only checked a few LLMs, and the conclusion may not generalize to all LLMs.

## Variability of Reasoning

We also want to check if the reasoning is consistent.

````python
df_identical = df_reasoning[df_reasoning.duplicated()]
print(df_reasoning.shape)
print(df_identical.shape)
````

As shown above, not only the responses but also the reasonings are identical with varied random_seed
